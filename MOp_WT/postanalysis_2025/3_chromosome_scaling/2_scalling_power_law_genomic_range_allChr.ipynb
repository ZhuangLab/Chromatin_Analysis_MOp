# Different genomic scales using power law function

```python
from scipy.spatial.distance import pdist, squareform
from scipy.stats import pearsonr, spearmanr
from astropy.convolution import convolve
from astropy.convolution import Gaussian2DKernel
from scipy.ndimage import gaussian_filter
from numpy import ma



def lighten_color(color, amount=0.5):
    """
    Lightens the given color by mixing it with white.
    Amount specifies the degree of lightening: 0 is the original color, 1 is white.
    
    Args:
    - color: The color to lighten. Can be specified as a hex string or a named color.
    - amount: The amount to lighten by. Range is 0 to 1, where 0 is no change and 1 is white.
    
    Returns:
    - The lightened color in hex format.
    """
    import matplotlib.colors as mc
    import colorsys
    
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])
```

```python
from scipy.optimize import curve_fit
def power_law(x, a, k):
    return a * np.power(x, k)
```

## short range

```python
distance_cutoff = [0,1e7]
# excluding ChrX and ChrY
sorted_chroms = [str(_ch) for _ch in range(1,20)]
```

### selected  cell types

```python
lighter_celltype_color_dict = {celltype: lighten_color(color, amount=0.35) for celltype, color in celltype_palette.items()}
sel_class_list = ['L2/3 IT', 'Pvalb','Oligo', 'Endo']


fig, ax =plt.subplots(1, len(sel_class_list), figsize = (1.*len(sel_class_list),1.2), dpi=150)

for _idx, _celltype in enumerate(sel_class_list):
    
    # init list to save dist and gdist
    gdist_entries_all = []
    median_entries_all = []
    
    for sel_chr in sorted_chroms:
        _chr_codebook = codebook_df.loc[codebook_df['chr']==sel_chr]
        _region_centers = np.sort([np.mean(np.array(_n.split(sel_chr+':')[1].split('-'), dtype=np.int32)) 
                                   for _n in _chr_codebook['name']])


        # get distance mat and gdist mat
        median_mat = class_2_medianDict[_celltype][f'cis_{sel_chr}']
        gdist_mat = squareform(pdist(_region_centers[:,np.newaxis]))

        gdist_entries = gdist_mat[np.triu_indices(len(gdist_mat),1)]
        median_entries = median_mat[np.triu_indices(len(median_mat),1)]
        
        # slice by genomic distance
        mask = (gdist_entries > distance_cutoff[0]) & (gdist_entries < distance_cutoff[1])
        gdist_entries = gdist_entries[mask]
        median_entries = median_entries[mask]
    
        valid_inds = np.isfinite(np.log(gdist_entries)) & np.isfinite(np.log(median_entries))
        gdist_entries = gdist_entries[valid_inds]
        median_entries = median_entries[valid_inds]
        
        gdist_entries_all.extend(gdist_entries)
        median_entries_all.extend(median_entries)

    gdist_entries_all = np.array(gdist_entries_all)
    median_entries_all = np.array(median_entries_all)

    # scatterplot
    ax[_idx].scatter(gdist_entries_all, median_entries_all, s=0.5,c=lighter_celltype_color_dict[_celltype], 
                         rasterized=True, alpha=1)
    
    ax[_idx].set_ylim([0.1,2.3])

    # calculate and plot power law 
    params, cov = curve_fit(power_law, gdist_entries_all, median_entries_all)
    a_fit, k_fit = params
    std_devs = np.sqrt(np.diag(cov))
    std_a, std_k = std_devs

    gdist_plot = np.linspace(gdist_entries_all.min(), gdist_entries_all.max(), 200)
    y_plot = a_fit * np.power(gdist_plot, k_fit)
    ax[_idx].plot(gdist_plot, y_plot, c='blue', linewidth=1., linestyle='--')


    # plotting params
    #ax[_idx].set_ylabel('Spatial distance (\u00B5m)', fontsize=7)
    #ax[_idx].set_xlabel('Genomic distance (bp)', fontsize=7)
    ax[_idx].tick_params(axis='both', which='both', labelsize = 6.5,
                     length=1, width=0.5, pad=1, 
                     direction='out', bottom=True, left=True
                    #labelleft=True, labelbottom=True,labelright=False, labeltop=False,
                       )

    ax[_idx].set_title(f'{_celltype} \n S={round(k_fit,2)} ± {round(std_k,2)}', fontsize=7)

    [i[1].set_linewidth(0.5) for i in ax[_idx].spines.items()]
        
    ax[_idx].set_ylim([0.1,2.3])
    ax[_idx].set_yticklabels([])
    ax[_idx].set_xticks([0*1e7, 0.5*1e7, 1*1e7])
    ax[_idx].set_xticklabels(['0Mb', ' ', '10Mb'])

    
plt.tight_layout()
plt.subplots_adjust(hspace=1)
plt.savefig(os.path.join(output_figure_folder,f'scaling_short_range_allChr_sel_celltypes.pdf'),
            transparent=True, bbox_inches='tight', pad_inches=0.2, dpi=300)
plt.show()
```
