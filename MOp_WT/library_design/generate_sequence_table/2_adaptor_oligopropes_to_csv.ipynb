#### CTP11 genome 3-color

```python
# bit from library design
adaptor_bit_folder = r'\\10.245.74.212\Chromatin_NAS_2\Chromatin_Libraries\CTP-11_brain\Summary_tables'
adaptor_bit_fname = os.path.join(adaptor_bit_folder, r'CTP11-mouse-genome-1000_combo_readouts.csv')
```

```python
adaptor_bit = pd.read_csv(adaptor_bit_fname)
adaptor_bit.tail()
```

```python
### get used adaptors in simialr format as RNA MERFISH
# 'Bit-1-RS0015_2xStv_82rc'
color_2_readout_dict = {'750':'_82rc','647':'_1rc','561':'_79rc'}
sel_type = 'c'
sel_type_when_dup = 'l'
_name = 'Genome chromatin adaptors'

used_adaptor_bit_dict = {'name':[],'sequence':[]}

for _bit in adaptor_bit['Bit'].tolist():
    readout_name = adaptor_bit[adaptor_bit['Bit']==_bit]['ReadoutName'].values[0]
    _sel_adaptor_df = merged_adaptors_df[merged_adaptors_df['name'].str.contains(readout_name)]

    cand_used_adaptor_color = []
    cand_used_adaptor_type = []
    for _ch in ['750', '647', '561']:
        _ch_adaptor_usage = adaptor_usage[_ch].tolist()
        _ch_adaptor_usage = [_a for _a in _ch_adaptor_usage if isinstance(_a, str)]
        _ch_adaptor_usage = [_a for _a in _ch_adaptor_usage if _a[0] in [sel_type, sel_type_when_dup]]
        _ch_adaptor_usage = [_a for _a in _ch_adaptor_usage if int(_a[1:])==_bit]
        # append all used adaptors (including failed)
        cand_used_adaptor_type.extend(_ch_adaptor_usage)
        cand_used_adaptor_color.extend([_ch]*len(_ch_adaptor_usage))
    
    # get the actual good adaptor used in decoding
    ## if only one hyb done
    if len(cand_used_adaptor_type)==1 and len(cand_used_adaptor_color)==1:
        _used_color = cand_used_adaptor_color[0]
        _used_adaptor_df = _sel_adaptor_df[_sel_adaptor_df['name'].str.contains(color_2_readout_dict[_used_color])]
        _used_adaptor_name = _used_adaptor_df['name'].values[0]
        used_adaptor_bit_dict['name'].append(f'Bit-{_bit}-{_used_adaptor_name}')
        used_adaptor_bit_dict['sequence'].append(_used_adaptor_df['sequence'].values[0])
        
    ### if more than one hyb done
    elif len(cand_used_adaptor_type)>1 and len(cand_used_adaptor_color) == len(cand_used_adaptor_type):
        _used_idx = np.where(np.array([_t[0] for _t in cand_used_adaptor_type])==sel_type_when_dup)[0]
        if len(_used_idx) == 1:
            _used_color = cand_used_adaptor_color[_used_idx[0]]
            _used_adaptor_df = _sel_adaptor_df[_sel_adaptor_df['name'].str.contains(color_2_readout_dict[_used_color])]
            _used_adaptor_name = _used_adaptor_df['name'].values[0]
            used_adaptor_bit_dict['name'].append(f'Bit-{_bit}-{_used_adaptor_name}')
            used_adaptor_bit_dict['sequence'].append(_used_adaptor_df['sequence'].values[0])
        else:
            used_adaptor_bit_dict['name'].append(f'Bit-{_bit}-Error')
            used_adaptor_bit_dict['sequence'].append('Error')
    else:
        used_adaptor_bit_dict['name'].append(f'Bit-{_bit}-Error')
        used_adaptor_bit_dict['sequence'].append('Error')

        
probes_df = pd.DataFrame(used_adaptor_bit_dict)
# check if there are any issues
print(len(probes_df[probes_df['name'].str.contains('Error')]))


csv_fname = os.path.join(output_sub_folder, f'{_name}.csv')
if not os.path.exists(csv_fname):
    print(f'Save {_name}')
    probes_df.to_csv(csv_fname, index=False)
else:
    print(f'Skip {_name}')
        
probes_df
```
